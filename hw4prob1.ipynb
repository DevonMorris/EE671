```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 2.5, 3, 5, 9])
y = np.array([-4.2, -5, 2, 1, 24.4])
```

```python
fig = plt.figure(1)
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
```

```python
ones = np.ones_like(x)
A = np.column_stack([ones, x])
c = la.inv(A.T.dot(A)).dot(A.T).dot(y)
```

```python
t = np.linspace(2, 9)
line = c[0] + c[1]*t
plt.scatter(x,y)
plt.plot(t,line)
plt.title("Linear Regression")
plt.show()
```

```python
W = np.eye(5)
for i in [1,2,3]:
    W[i,i] = 0
W
```

```python
c = la.inv(A.T.dot(W).dot(A)).dot(A.T.dot(W)).dot(y)
```

```python
t = np.linspace(2, 9)
line = c[0] + c[1]*t
plt.scatter(x,y)
plt.plot(t,line)
plt.title("Linear Regression with Weighting")
plt.show()
```
