```python
%matplotlib inline
import numpy as np
import numpy.linalg as la
import scipy.io
import matplotlib.pyplot as plt
```

```python
mat = scipy.io.loadmat('hw3prob5.mat')

A = mat['A']
x = mat['x']
x.shape
plt.imshow(x, cmap='gray')
plt.show()
```

```python
B = A.reshape(-1, A.shape[-1])
y = x.reshape(-1)
```

```python
z = y - B.dot(la.inv(B.T.dot(B)).dot(B.T).dot(y))
```

```python
img = z.reshape(256,256)
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

The image appears degraded after being decoded because it has been projected on the orthogonal subspace to the space spaned by the 20 images. Basically that means we've lost a subspace of dimension 20! Since that subspace contained semantically important information the image appears degraded.
